# Group by and Aggregating

```python
df3 = pd.read_excel(r'Downloads\Employee Data.xlsx')
```

```python
df3
```

```python
group_frame = df3.groupby('Age Group')
```

```python
group_frame.count()
```

```python
dfn = df3[['Age Group','Salary']]
```

```python
group_f = dfn.groupby('Age Group')
```

```python
group_f
```

```python
group_f.mean()
```

```python
group_frame.min()
```

```python
group_frame.max()
```

```python
group_frame.sum()
```

```python
group_frame.agg({'Salary':['mean','min','max','sum']})
```

```python
df3.groupby(['Age Group','Department']).max()
```
